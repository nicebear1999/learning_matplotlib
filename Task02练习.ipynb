请自行创建数据，画出包含6个子图的线图，要求：
子图排布是 2 * 3 （2行 3列）；
线图可用教程中line2D方法绘制；
需要设置每个子图的横坐标和纵坐标刻度； 并设置整个图的标题，横坐标名称，以及纵坐标名称

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

fig=plt.figure()
for i in range(1,7):
    x = range(0,5)
    y = np.random.randint(0,11,5)
    ax=fig.add_subplot(2,3,i)
    ax.plot(x,y)
    plt.xticks(x)
    plt.yticks(range(0,11,2))
```
